# Convolutions

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, color
from skimage.transform import resize

# Load a sample image from scikit-image and convert it to grayscale
image = color.rgb2gray(data.coffee())

# Resize image for faster processing
image = resize(image, (128, 128))

# Define filters
edge_filter = np.array([
    [-1, -1, -1],
    [-1, 8, -1],
    [-1, -1, -1]
])

blur_filter = np.array([
    [1/9, 1/9, 1/9],
    [1/9, 1/9, 1/9],
    [1/9, 1/9, 1/9]
])

# Function to perform convolution
def convolve2d(image, filter):
    filter_size = filter.shape[0]
    result_size = image.shape[0] - filter_size + 1
    result = np.zeros((result_size, result_size))
    
    for i in range(result_size):
        for j in range(result_size):
            result[i, j] = np.sum(image[i:i+filter_size, j:j+filter_size] * filter)
    
    return result

# Apply filters to the image
output_edge = convolve2d(image, edge_filter)
output_blur = convolve2d(image, blur_filter)

# Plot the original image and the filtered images
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title("Edge Detection")
plt.imshow(output_edge, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title("Blurred Image")
plt.imshow(output_blur, cmap='gray')
plt.axis('off')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, color

# Load a sample image from scikit-image and convert it to grayscale
image = data.brick()

# Resize image for faster processing
image = resize(image, (100, 100))

# Define Sobel filters for edge detection
sobel_vertical = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
])

sobel_horizontal = np.array([
    [-1, -2, -1],
    [0, 0, 0],
    [1, 2, 1]
])

# Function to perform convolution
def convolve2d(image, filter):
    filter_size = filter.shape[0]
    result_size = image.shape[0] - filter_size + 1
    result = np.zeros((result_size, result_size))
    
    for i in range(result_size):
        for j in range(result_size):
            result[i, j] = np.sum(image[i:i+filter_size, j:j+filter_size] * filter)
    
    return result


# Apply filters to the image
output_vertical = convolve2d(image, sobel_vertical)
output_horizontal = convolve2d(image, sobel_horizontal)

# Round the values in the result matrices
output_vertical_rounded = np.round(output_vertical)
output_horizontal_rounded = np.round(output_horizontal)
image_rounded = np.round(image, 1) #precision just detates just how defined those features are going to be. 

# Plot the filters and their effect on the image
plt.figure(figsize=(15, 10))

# Visualize the filters
plt.subplot(2, 3, 1)
plt.title("Sobel Vertical Filter")
plt.imshow(sobel_vertical, cmap='gray')
plt.colorbar()
plt.axis('off')

plt.subplot(2, 3, 2)
plt.title("Sobel Horizontal Filter")
plt.imshow(sobel_horizontal, cmap='gray')
plt.colorbar()
plt.axis('off')

# Visualize the original image
plt.subplot(2, 3, 3)
plt.title("Original Image")
plt.imshow(image_rounded,cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar()
plt.axis('off')

# Visualize the rounded filtered images with colormap
plt.subplot(2, 3, 4)
plt.title("Rounded Vertical Edges")
plt.imshow(output_vertical_rounded, cmap='coolwarm', vmin=-4, vmax=4)
plt.colorbar()
plt.axis('off')

plt.subplot(2, 3, 5)
plt.title("Rounded Horizontal Edges")
plt.imshow(output_horizontal_rounded, cmap='coolwarm', vmin=-4, vmax=4)
plt.colorbar()
plt.axis('off')

plt.tight_layout()
plt.show()

```
